### Usando a API de streaming do Twitter
A maioria dos usuários provavelmente já coletou seus dados do Twitter para realizar algumas análises, mas segue abaixo um exemplo simples de como coletar dados de tweets de streaming que são coletados usando um filtro, usando uma hashtag específica.

O código abaixo é um exemplo simples de como coletar tweets com a hashtag *#nerd*, imprimir os dados do tweet e salvá-lo em um arquivo.

```python
import re
import datetime
import tweepy
from __future__ import absolute_import, print_function
```

```python
# Go to http://apps.twitter.com and create an app.
# The consumer key and secret will be generated for you after
consumer_key="9mmvxstYsxZJge3XAztvDYEX1"
consumer_secret="1q0tYBnBKK0dBZ1ENOFDNDpmrcDOfDToLiMPsK7Lyv6TyPWhes"

# After the step above, you will be redirected to your app's page.
# Create an access token under the the "Your access token" section
access_token="110406901-2WAs5DAxu0h5fNKQdGdKIWQXlcVq6Jlpg3npEzku"
access_token_secret="GhVbT7e8vmD0I9bwWY2mh6npcqj1AqqQg8qAWXyUelZ4K"

# Create variables to use for output file and tweet filter
hashtag = "climatechange"
date_time_temp = str(datetime.datetime.now())

# Replace all characters except letters and numbers with "_" for filename
current_date_time = re.sub('[^a-zA-Z0-9]','_', date_time_temp)
file_out = open(hashtag + "_" + current_date_time + ".json", 'a')
```

```python
# Define the Stream Listener
class StdOutListener(tweepy.Stream):
    """ A listener handles tweets that are received from the stream.
    This is a basic listener that just prints received tweets to stdout.

    """
    def on_data(self, data):
        print(data) # Print output to console
        file_out.write(str(data)) # Write output to file
        return True

    def on_error(self, status):
        print(status)

# Run the main program and collect tweets with hashtag "climatechange"
if __name__ == '__main__':
    # Initialize instance of the subclass
    stream = StdOutListener(
      consumer_key, consumer_secret,
      access_token, access_token_secret
    )

    stream.filter(track=['climatechange'])
file_out.close()
```
